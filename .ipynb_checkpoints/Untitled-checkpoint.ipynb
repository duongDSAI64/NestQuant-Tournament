```python
import os
import pyarrow.parquet as pq
import pandas as pd

# Specify the directory containing the Parquet files
directory = 'Data/Extracted/BTCUSDT/Data'

# Get a list of Parquet files in the directory
parquet_files = [file for file in os.listdir(directory) if file.endswith('.parquet')]

# Read and concatenate the Parquet files
data_frames = []
for file in parquet_files:
    file_path = os.path.join(directory, file)
    table = pq.read_table(file_path)
    df = table.to_pandas()
    data_frames.append(df)

# Concatenate the data frames
concatenated_df = pd.concat(data_frames)

# Print the concatenated data frame
concatenated_df
```

```python
label_directory = 'Data/Extracted/BTCUSDT/Label'
label_parquet_files = [file for file in os.listdir(label_directory) if file.endswith('.parquet')]
# Read and concatenate the Parquet files
label_data_frames = []
for file in label_parquet_files:
    file_path = os.path.join(label_directory, file)
    table = pq.read_table(file_path)
    df = table.to_pandas()
    label_data_frames.append(df)

# Concatenate the data frames
label_df = pd.concat(label_data_frames)

# Print the concatenated data frame
label_df
```

```python
result = pd.merge(concatenated_df, label_df, on='OPEN_TIME', how='outer')
```
